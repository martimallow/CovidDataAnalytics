```python
df_class = pd.read_csv('../data/act.csv', index_col=0)
```

```python
def label_V4013(row):
    try:
        cod = float(row)
        if math.isnan(cod):
            return cod
        cod = int(cod)
        return df_class.loc[cod]['essential']
    except ValueError:
        print(row)
        return row
```

```python
def label_V4013_group(row):
    try:
        cod = float(row)
        if math.isnan(cod):
            return cod
        cod = int(cod)
        return df_class.loc[cod]['group']
    except ValueError:
        print(row)
        return row
```

```python
#Create column 'essential'
df['essential'] = df['V4013'].map(label_V4013)
```

```python
#Create column 'group'
df['group'] = df['V4013'].map(label_V4013_group)
```

```python
#Comparing results
print(df['V4013'].tail(20))
print('--')
print(df['essential'].tail(20))
print('--')
print(df['group'].tail(20))
```
