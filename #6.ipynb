<h1>Por tipo de area<h1>

```python
#V1023: Tipo de area
# 1: Capital
# 2: Resto da Região Metropolitana
# 3: Resto da RIDE (Região Integrada de Desenvolvimento Econômico, excluindo a capital)
# 4: Resto da UF  (Unidade da Federação, excluindo a região metropolitana e a RIDE)
#VD4002: Pessoas ocupadas
```

```python
regiao2019_1 = p2019_1[['V1023', 'VD4002']][p2019_1['VD4002'] == 1].groupby('V1023').sum()
regiao2019_1.loc['5'] = regiao2019_1.iloc[0:2].sum(numeric_only=True)
regiao2019_1['Trimestre'] = '2019-1'
regiao2019_1
```

```python
regiao2019_2 = p2019_2[['V1023', 'VD4002']][p2019_2['VD4002'] == 1].groupby('V1023').sum()
regiao2019_2.loc['5'] = regiao2019_2.iloc[0:2].sum(numeric_only=True)
regiao2019_2['Trimestre'] = '2019-2'
regiao2019_2
```

```python
regiao2019_3 = p2019_3[['V1023', 'VD4002']][p2019_3['VD4002'] == 1].groupby('V1023').sum()
regiao2019_3.loc['5'] = regiao2019_3.iloc[0:2].sum(numeric_only=True)
regiao2019_3['Trimestre'] = '2019-3'
regiao2019_3
```

```python
regiao2019_4 = p2019_4[['V1023', 'VD4002']][p2019_4['VD4002'] == 1].groupby('V1023').sum()
regiao2019_4.loc['5'] = regiao2019_4.iloc[0:2].sum(numeric_only=True)
regiao2019_4['Trimestre'] = '2019-4'
regiao2019_4
```

```python
regiao2020_1 = p2020_1[['V1023', 'VD4002']][p2020_1['VD4002'] == 1].groupby('V1023').sum()
regiao2020_1.loc['5'] = regiao2020_1.iloc[0:2].sum(numeric_only=True) 
regiao2020_1['Trimestre'] = '2020-1'
regiao2020_1
```

```python
regiao_frames = [regiao2019_1,regiao2019_2,regiao2019_3,regiao2019_4,regiao2020_1]
regiao = pd.concat(regiao_frames)
```

```python
sns.barplot(x=regiao.index, y='VD4002', hue='Trimestre', data=regiao)
plt.title('Ocupados por tipo de área')
plt.ylabel('Total de ocupados')
plt.xlabel('Tipo de área')
plt.savefig('ocupados_area.png', transparent=True)
plt.show()
```

<h1>Por raça<h1>

```python
#V2010: Cor ou raça
# 1	Branca
# 2	Preta
# 3	Amarela
# 4	Parda 
# 5	Indígena
# 9	Ignorado

#VD4002: Condição de ocupação
```

```python
raca2019_1 = p2019_1[['V2010','VD4002']][p2019_1['VD4002'] == 1].groupby('V2010').sum()
raca2019_1['Trimestre'] = '2019-1'
raca2019_1
```

```python
raca2019_2 = p2019_2[['V2010','VD4002']][p2019_2['VD4002'] == 1].groupby('V2010').sum()
raca2019_2['Trimestre'] = '2019-2'
raca2019_2
```

```python
raca2019_3 = p2019_3[['V2010','VD4002']][p2019_3['VD4002'] == 1].groupby('V2010').sum()
raca2019_3['Trimestre'] = '2019-3'
raca2019_3
```

```python
raca2019_4 = p2019_4[['V2010','VD4002']][p2019_4['VD4002'] == 1].groupby('V2010').sum()
raca2019_4['Trimestre'] = '2019-4'
raca2019_4
```

```python
raca2020_1 = p2020_1[['V2010','VD4002']][p2020_1['VD4002'] == 1].groupby('V2010').sum()
raca2020_1['Trimestre'] = '2020-1'
raca2020_1
```

```python
raca_frames = [raca2019_1,raca2019_2,raca2019_3,raca2019_4,raca2020_1]
raca = pd.concat(raca_frames)
```

```python
sns.barplot(x=raca.index, y='VD4002', hue='Trimestre', data=raca)
plt.title('Ocupados por cor ou raça')
plt.ylabel('Total de ocupados')
plt.xlabel('Cor ou raça')
plt.savefig('ocupados_raca', transparent=True)
plt.show()
```

```python
#Criando a categoria negros (pretos e pardos)
negros = raca[raca.index.isin([2,4])].groupby('Trimestre', as_index=False).sum()
negros['V2010'] = 6
negros = negros.set_index('V2010')
negros
```

```python
#Selecionando apenas negros e brancos
bw = [raca[raca.index == 1], negros]
black_white = pd.concat(bw).rename(index={1 : 'Brancos', 6 : 'Negros'})
black_white
```

```python
sns.barplot(x=black_white.index, y='VD4002', hue='Trimestre', data=black_white)
plt.title("Brancos e Negros ocupados")
plt.ylabel("Total de ocupados")
plt.xlabel("cor ou raça")
#plt.savefig('ocupados_b&w.png', transparent=True)
plt.show()
```
